```python
import numpy as np
import matplotlib.pyplot as plt
import librosa
from IPython.display import Audio
import os
```

```python
y1, sr1 = librosa.load("./data/dummy_conversation.weba", sr=None)
y2, sr2 = librosa.load("./data/dummy_conversation_mic.weba", sr=None)
```

```python
plt.figure(figsize=(16, 6))

plt.subplot(121)
librosa.display.waveshow(y1, sr=sr1, color="blue")
plt.ylabel("Amplitude")
plt.xlabel("Time (secs)")
plt.title("Audio Waveforms")

plt.subplot(122)
librosa.display.waveshow(y2, sr=sr2, color="red")
plt.ylabel("Amplitude")
plt.xlabel("Time (secs)")
plt.title("Audio Waveforms")
```

```python
fft1 = np.fft.fft(y1)
mag1 = np.abs(fft1)[:len(fft1) // 2]
freq1 = np.linspace(0, sr1 / 2, len(mag1))

fft2 = np.fft.fft(y2)
mag2 = np.abs(fft2)[:len(fft2) // 2]
freq2 = np.linspace(0, sr2 / 2, len(mag2))

plt.figure(figsize=(16, 6))

plt.subplot(121)
plt.plot(freq1, mag1)

plt.subplot(122)
plt.plot(freq2, mag2)
```
